```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("all_questions.csv")
```

```python
import math
def myhot(Qviews, Qanswers, Qscore, Ascore):
    return (math.log10(Qviews) * 4 + float(Qanswers * Qscore / 5) + float(Ascore))
```

```python

import pymysql

def connectDB():
    db = pymysql.connect(host="localhost", user="root", password="", database="deeplearning")
    print("connected...")
    return db

def queryDb(db, sql):
    cursor = db.cursor()
    cursor.execute(sql)
    results = cursor.fetchone()
    return results
```

```python
db = connectDB()
df['hot_score'] = 0
for index, row in df.iterrows():
    Qviews = row['view_count']
    Qanswers = row['answer_count']
    Qscore = row['score']
    sql = "select sum({}) from {} where parent_id = {}".format("score", "answers", row['post_id'])
    result = queryDb(db, sql)
    Ascore = 0
    if(result[0] is not None):
        Ascore = result[0]
    df.at[index, 'hot_score'] = myhot(Qviews, Qanswers, Qscore, Ascore)
db.close()
```

```python
hotquestion = df[df['hot_score']>=50]
hotquestion
```

```python
hotquestion.to_csv("hotquestion.csv")
```
